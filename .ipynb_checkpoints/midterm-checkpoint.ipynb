The Relative Rate of Growth is given formula 1/P.dP/dt, we only calculate these values from when there was first death case ever. 

```python
death_RG = df["Deaths"]
rate_of_growth = []
for i in range(94, 123):
    if i == 94:
        rate_growth = ((death_RG[i+1] - death_RG[i])/death_RG[i])
    elif i == 122 :
        rate_growth = ((death_RG[i] - death_RG[i-1])/death_RG[i])
    else:
        rate_growth = ((death_RG[i+1] - death_RG[i-1])/ (2*death_RG[i]))
    rate_of_growth.append(round(rate_growth,3))
```

```python
rate_of_growth_array = np.asarray(rate_of_growth)
```

```python
rate_of_growth_array
```

```python
df['Rate'] = rate_of_growth
```

```python
df
```

```python
x = df['Date']
y = df['Deaths']
plt.figure(figsize=(20,10))
plt.plot(x, y, 'bo')
ax = plt.gca()
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total deaths', fontsize=14)
plt.xticks(fontsize=10, rotation = 10)
plt.yticks(fontsize=10)
ax.xaxis.set_major_locator(mdates.DayLocator(interval=10))
ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gcf().autofmt_xdate()
plt.title('Total Deaths in Spain up to 2020-05-01', fontsize=20)
plt.show()
```

From equation we can implement that k = 0.1154 and K = 26150 => A = 1.6142 => P has formula below :

```python
from math import e 
x = df['Date']
y = df['Deaths']
p = np.linspace(1, 29, 29)

plt.figure(figsize=(20,10))
plt.plot(p, y, 'bo')
ax = plt.gca()
plt.xlabel('Date', fontsize=14)
plt.ylabel('Total deaths', fontsize=14)
#plt.xticks(fontsize=10, rotation = 10)
#plt.yticks(fontsize=10)
#ax.xaxis.set_major_locator(mdates.DayLocator(interval=10))
#ax.xaxis.set_major_formatter(mdates.DateFormatter('%d-%m-%Y'))
plt.gcf().autofmt_xdate()

#plt.figure(figsize=(20,10))
plt.plot(p,(26150/(1+1.6142*e**(-0.1154*p))))
plt.title('Difference between observed and predicted Total Deaths in Spain up to 2020-05-01', fontsize=20)

plt.show()
```

```python
predicted = (26150/(1+1.6142*e**(-0.1154*p)))
for i in range(len(predicted)):
    predicted[i] = round(predicted[i],0)
```

```python
df['Predicted'] = predicted
```

```python
df
```
